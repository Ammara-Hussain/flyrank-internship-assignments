```python
import pandas as pd

# Paste your copied Raw URL here
raw_csv_url = 'https://raw.githubusercontent.com/Ammara-Hussain/flyrank-internship-assignments/refs/heads/main/data/raw/content_refresh_anonymized.csv'

# Load the dataset directly into a Pandas DataFrame
df = pd.read_csv(raw_csv_url)

# Inspect your dataset
print("Dataset Shape:", df.shape)
df.head()
```

```python
# Print all column names to see what metrics we have
print(df.columns.tolist())
```

```python
# Check how many missing values each column has
missing_summary = df.isnull().sum()
print("Columns with missing values:\n", missing_summary[missing_summary > 0])
```

```python
# Create a copy to keep raw data intact
df_cleaned = df.copy()

# Fill missing numerical values with 0 or column median
df_cleaned['search_volume'] = df_cleaned['search_volume'].fillna(0)
df_cleaned['competition'] = df_cleaned['competition'].fillna(df_cleaned['competition'].median())

# Calculate a Priority Score (Example logic based on low competition + high volume)
# (Adjust formula based on your exact column names)
df_cleaned['priority_score'] = (
    (df_cleaned['search_volume'] / df_cleaned['search_volume'].max()) * 0.6 +
    ((1 - df_cleaned['competition'])) * 0.4
)

# Rank from highest priority (1) to lowest
df_cleaned['priority_rank'] = df_cleaned['priority_score'].rank(ascending=False)

# Extract top 20 pages that need review/refresh
top_20 = df_cleaned.sort_values('priority_rank').head(20)
top_20[['content_id', 'client_id', 'search_volume', 'competition', 'priority_score', 'priority_rank']]
```
